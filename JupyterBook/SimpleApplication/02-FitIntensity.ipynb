## 2. Fitting kappa, vsini, vmac, and vrad

The `fitparams.py` module contains a function that can fit number or combination of the above parameters for any number of stars. For this example we will keep fit all four parameters for a single star. For any parameter not being fit, the current entry in the param object will be used.

We must first set up our param dictionary and DataPacket. Here, we just read in the param file we created earlier and reading the data packet. We then must update the `vrad` option to be a list with one value per observation.

````python
param_initial = rav.params.read_parameters('ExampleOutput/02-FitIntensity/FitI_param_initial.json')
DataPacket = rav.data.read_packet('ExampleOutput/01-DataSetup/ExamplePacket.h5')

#updating the vrad entry
nobs=DataPacket.nobs
param_initial['general']['vrad']=[0]*nobs
#printing the initial parameters
param_initial.pprint()
````

The `fitparams` function uses scipy's `minimize` function to perform the fit. The function requires the following inputs:
* `param_to_fit` - a list of lists containing the names of parameters to fit. i.e. for 1 star [['logkappa','vmac','vrad','vsini']]
* `parameters` - a list containing a param object (in this case `param_initial`)
* `lsds` - a list containing 3 lists, a velocity array, a specI array, and a specSigI array for each obseravion.
* `guess` - a list of lists containing the guess values for each of those parameters in the same order as `param_to_fit`. for the `vrad` entry you must enter another list with 1 value per observation
* `bounds` - a list of lists containing a tuple with the lower and upper bounds for each `param_to_fit`

:::{admonition} TODO
:class: warning
Fit is done on the Packet.scaled -- should be done on the Packet.cutfit ?
:::

::::{margin}
:::{tip} The order of entries in the `param_to_fit`, `guess`, and `bounds` entries doesn't matter as long as the order is the same for them all. i.e. `param_to_fit` = [['logkappa','vmac','vrad','vsini']] or [['vmac','vrad','vsini','logkappa']]
:::
::::


````python
param_to_fit = [['logkappa','vmac','vsini','vrad']]
guess = [[-0.1,30,150,[0]*nobs]]
bounds = [[(None,None),(0,50),(5,500),[(-100,100)]*nobs]]
parameters=[param_initial]

vels=[]
specIs=[]
specIs_err=[]
for o in range(nobs):
    vels.append(DataPacket.scaled.lsds[o].vel)
    specIs.append(DataPacket.scaled.lsds[o].specI)
    specIs_err.append(DataPacket.scaled.lsds[o].specSigI)

lsds=[vels,specIs,specIs_err]
````

We can then fit the data by calling the `fitting` routine. There are two outputs; the scipy's `minimize` function output, and a list of param objects in the same format as the `parameters` entry which contains the updated fit values.

````python
res,parameters_best=rav.fitparams.fitting(param_to_fit,parameters,lsds,guess,bounds)
parameters_best[0].pprint()
````

We can make a graph of the data and the best fit

````python
# Make a figure and plot the scaled LSD Stokes I profiles
fig, ax = plt.subplots(1,3,figsize=(12,4))

param_best=parameters_best[0]
model = rav.diskint.analytical(param_best)
for o in range(nobs):
    ax[o].plot(vels[o],specIs[o])
    ax[o].plot(model['vel']+param_best['general']['vrad'][0],model['flux'],c='k',ls='dotted',label='Analytical')
    ax[o].set_title('Observation {}'.format(o+1))


ax[0].legend(loc=0)
plt.show()
````

As a second example, we can perform a fit of `logkappa`, `vmac`, and `vrad` without fitting `vsini`. Again, this can also be done for any combination of the four parameters.

````python
param_initial['general']['vsini']=185 #since vsini is not being fit, whatever value we set here will be assumed to be true

param_to_fit = [['logkappa','vmac','vrad']]
guess = [[-0.1,30,[0]*nobs]]
bounds = [[(None,None),(0,50),[(-100,100)]*nobs]]
parameters=[param_initial]

vels=[]
specIs=[]
specIs_err=[]
for o in range(nobs):
    vels.append(DataPacket.scaled.lsds[o].vel)
    specIs.append(DataPacket.scaled.lsds[o].specI)
    specIs_err.append(DataPacket.scaled.lsds[o].specSigI)

lsds=[vels,specIs,specIs_err]

res,parameters_best=rav.fitparams.fitting(param_to_fit,parameters,lsds,guess,bounds)
parameters_best[0].pprint()
````

and now we can plot this fit

````python
# Make a figure and plot the scaled LSD Stokes I profiles
fig, ax = plt.subplots(1,3,figsize=(12,4))

param_best=parameters_best[0]
model = rav.diskint.analytical(param_best)
for o in range(nobs):
    ax[o].plot(vels[o],specIs[o])
    ax[o].plot(model['vel']+param_best['general']['vrad'][0],model['flux'],c='k',ls='dotted',label='Analytical')
    ax[o].set_title('Observation {}'.format(o+1))


ax[0].legend(loc=0)
plt.show()
````
